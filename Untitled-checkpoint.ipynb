```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
from scipy.stats import linregress
import numpy as np
from collections import Counter

#Import and read csv file

spotify_file = "../Resources/DataSpotify.csv"

spotify_df = pd.read_csv(spotify_file)

spotify_df.head()
```

```python
# Remove columns 
clean_df = spotify_df.drop(['Unnamed: 0', 'key', 'mode', 'time_signature', 'explicit', 'track_id', 'art_id', 'tempo', 'duration_ms', 'followers', 'genres'], axis=1)
clean_df
```

```python
#Rename columns
renamed_df = clean_df.rename(columns={"poistion":"Position", "track_name":"Song", "artist":"Artist", "popularity":"Popularity","danceability":"Danceability", "energy":"Energy", "loudness":"Loudness", "speechiness":"Speechiness", "instrumentalness":"Instrumentalness", "liveness":"Liveness", "valence":"Valence", "album": "Album", "year":"Year"})
renamed_df
```

```python
energy_df = renamed_df.groupby(by="Year")['Energy'].describe()

energy_df
```

```python
renamed_df.plot(kind="scatter", x="Year", y="Energy", figsize=(8,8),
              title="Top 100 Energy Value")
plt.savefig("energy_scatter.png", bbox_inches = 'tight')
plt.show()
```

```python
renamed_df.boxplot(column='Energy', by='Year', grid=False, rot=90, fontsize=15)
title_boxplot = 'Energy Outliers'
plt.title( title_boxplot )
plt.suptitle('')
plt.savefig("energy_box.png", bbox_inches = 'tight')
plt.show()
```

```python
mean_energy = energy_df['mean'].tolist()
x_axis = np.arange(len(mean_energy))

plt.title("Mean Energy Over Time")
plt.xlabel("Year")
plt.ylabel("Energy")

tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022])
plt.savefig("energy_bar.png", bbox_inches = 'tight')
plt.bar(x_axis, mean_energy, color='b', alpha=0.5, align="center");
```

```python
Results:Energy value is not related to a song popularity. In the top 100 over the years we can find songs with 
    high and low values of energy. Over the last 10 years the average of the energy value of the songs decreased.
    In 2020, the average of the nergy value of the top 100 was the lowest and the max value was also the lowest 
    on the last 10 years. 
```
