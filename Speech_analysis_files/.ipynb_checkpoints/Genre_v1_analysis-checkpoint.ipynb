```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import dataframe_image as dfi
```

```python
Final_df =pd.read_csv('Final_Data_2012_2022.csv')
```

```python
df2 = Final_df.groupby('year')['gen_1'].agg(['count'])
df2
```

```python
p_table = pd.pivot_table(Final_df, columns =['gen_1'], values = ['year'], aggfunc = 'count')
p_table
```

```python
p_style = p_table.style.background_gradient(cmap='PuRd', axis =None, subset =['Country', 'Dance', 'EDM', 'Folk', 'Hip Hop', 'House', 'Indie', 'Latin', 'None', 'Other', 'Pop', 'R&B', 'Rap', 'Reggae', 'Rock', 'Soul'])
p_style
```

```python
dfi.export(p_style, 'gen_all.png')
```

```python
gen_list =Final_df['gen_1'].unique().tolist()
gen_abc = sorted(gen_list)
gen_abc
```

```python
gen_size = Final_df.groupby(by='gen_1')['track_id'].count().tolist()
gen_size
```

```python
colour = ['LightGrey', 'Thistle', 'LavenderBlush', 'Orchid', 'PaleVioletRed', 'PeachPuff', 'Gainsboro', 'Plum', 'Lavender', 'Violet', 'LightPink', 'MistyRose','HotPink', 'Pink', 'RosyBrown', 'LightSlateGray']

gen_pie = plt.pie(gen_size, labels=gen_abc,
        autopct="%1.1f%%", shadow=True, colors = colour, startangle=140, pctdistance=0.9, labeldistance=1.1, radius = 3);

plt.savefig("Gen_pie.png", bbox_inches = 'tight')
plt.show()
```

```python
gen_pivot = Final_df.pivot_table(index='year', 
               columns='gen_1',
               aggfunc='size',
               fill_value=0)
```

```python
gen_style = gen_pivot.style.background_gradient(cmap='PuRd', axis =None, subset =['Country', 'Dance', 'EDM', 'Folk', 'Hip Hop', 'House', 'Indie', 'Latin', 'None', 'Other', 'Pop', 'R&B', 'Rap', 'Reggae', 'Rock', 'Soul'])
gen_style
```

```python
dfi.export(gen_style, 'Gen_byyear.png')
```

```python
gen_pivot2 = pd.merge(gen_pivot, df2, on='year')
```

```python
gen_pivot2
```

```python
gen_piv_all = pd.concat([gen_pivot2, p_table])
```

```python
count = gen_pivot2['count'].tolist()
dance_list = gen_pivot2['Dance'].tolist()
edm_list = gen_pivot2['EDM'].tolist()
hh_list = gen_pivot2['Hip Hop'].tolist()
pop_list = gen_pivot2['Pop'].tolist()
rap_list = gen_pivot2['Rap'].tolist()
```

```python
#bananasplit = [int(b) / int(m) for b,m in zip(banana, monkey)]

Dance_per = [round(int(d)/int(c)*100,2) for d,c in zip(dance_list, count)]
edm_per = [round(int(e)/int(c)*100,2) for e,c in zip(edm_list, count)]
hh_per = [round(int(h)/int(c)*100,2) for h,c in zip(hh_list, count)]
pop_per = [round(int(p)/int(c)*100,2) for p,c in zip(pop_list, count)]
rap_per = [round(int(r)/int(c)*100,2) for r,c in zip(rap_list, count)]
```

```python
gen_pivot2["Dance%"] = Dance_per
gen_pivot2["EDM%"] = edm_per
gen_pivot2["Hip Hop%"] = hh_per
gen_pivot2["Pop%"] = pop_per
gen_pivot2["Rap%"] = rap_per
gen_pivot2
```

```python
year = ['2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022',]
dance = gen_pivot2['Dance%'].tolist()
edm = gen_pivot2['EDM%'].tolist()
hh = gen_pivot2['Hip Hop%'].tolist()
pop = gen_pivot2['Pop%'].tolist()
rap = gen_pivot2['Rap%'].tolist()
```

```python
dance_p, = plt.plot(year, dance, color ='Thistle', label = "Dance")
edm_p, = plt.plot(year, edm, color ='LavenderBlush', label = "EDM")
hh_p, = plt.plot(year, hh, color ='PaleVioletRed', label = "Hip Hop")
pop_p, = plt.plot(year, pop, color ='LightPink', label = "Pop")
rap_p, = plt.plot(year, edm, color ='HotPink', label = "Rap")
plt.title("Top 5 Generes Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Tracks")
plt.legend(handles=[dance_p, edm_p, hh_p, pop_p, rap_p], bbox_to_anchor=(1, 1))
plt.savefig("Gen_time.png", bbox_inches = 'tight')
plt.show()
```
