```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import dataframe_image as dfi
import scipy.stats as st
from scipy.stats import linregress
```

```python
Track22_df =pd.read_csv('2022Year_final.csv')
Track22_df
```

```python
df=pd.read_csv
df1=Track22_df.iloc[:93,[1,3,5]]
df1
```

```python
artist=df1['artist'].tolist()
followers=df1['followers'].tolist()
y=range(len(artist))

plt.figure(figsize=(15,4))
plt.bar(y,followers)
plt.xticks(y,artist,rotation=90);
```

```python
plt.scatter(df1.iloc[:,0],df1.iloc[:,2])

#Create Scatter plot of all Songs populatiry vs speechiness 

title_label = "Followers vs Top Tracks Poistion"
x_label = 'Top Hits Poistion'
y_label = 'Number of Followers'

  # Compute linear regression
(slope, intercept, rvalue, pvalue, stderr) = linregress(df1.iloc[:,0], df1.iloc[:,2])
regress_values = df1.iloc[:,0] * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

    # Plot
plt.scatter(df1.iloc[:,0],df1.iloc[:,2],marker="s", facecolors="lightblue", edgecolors="black", alpha = 0.7)
plt.plot(df1.iloc[:,0],regress_values,"r-")
plt.xlabel(x_label)
plt.ylabel(y_label)
plt.title(title_label)
print(f"The r-value is: {rvalue**2}")
correlation = st.pearsonr(df1.iloc[:,0],df1.iloc[:,2])
print(f"The correlation between both factors is {round(correlation[0],2)}")
print(line_eq)

plt.savefig("Follow_scat1.png", bbox_inches = 'tight')
plt.show()
```
